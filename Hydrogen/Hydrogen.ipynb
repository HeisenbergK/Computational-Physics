```python
from Hydrogen import Hydrogen
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nmax=3

plt.style.use('ggplot')
NUM_COLORS = 12
cm = plt.get_cmap('tab20')

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_prop_cycle(color=[cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])

hydro = Hydrogen(nmax, ax)
hydro.FindAllBound()
hydro.PlotAllStates()

plt.xlim([0,20])
plt.ylim([-0.25,1])
plt.legend(ncol=3)
plt.xlabel(r"r$\left(\frac{\hbar{}}{m_{e}c\alpha{}}\right)$")
plt.ylabel(r"rR(r)$\left(\frac{\hbar{}}{m_{e}c\alpha{}}\right)^{\frac{1}{2}}$")
plt.title(f"First {len(hydro.bnd)} states of the Hydrogen Atom (up to n={nmax})")
```
